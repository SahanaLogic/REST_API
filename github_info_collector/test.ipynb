```python
import os
import requests

GITHUB_TOKEN = os.environ.get("GITHUB_TOKEN")
if not GITHUB_TOKEN or GITHUB_TOKEN == "ghp_your_token_here":
    print("⚠️  GITHUB_TOKEN is not set. Run cell 4 first with your real token.")
else:
    headers = {
        "Authorization": f"Bearer {GITHUB_TOKEN}",
        "Accept": "application/vnd.github+json",
        "X-GitHub-Api-Version": "2022-11-28",
    }

    # 1) Get authenticated user
    user_resp = requests.get("https://api.github.com/user", headers=headers)
    user_resp.raise_for_status()
    user = user_resp.json()
    print("Logged in as:", user["login"])

    # 2) List user repos (first 5)
    repos_resp = requests.get("https://api.github.com/user/repos", headers=headers)
    repos_resp.raise_for_status()
    repos = repos_resp.json()

    for repo in repos[:5]:
        print(f"- {repo['name']} (private={repo['private']})")
```

```python
import os
import requests

# ---------- GitHub ----------
GITHUB_TOKEN = os.environ.get("GITHUB_TOKEN")
if not GITHUB_TOKEN:
    raise RuntimeError("Set GITHUB_TOKEN first")

gh_headers = {
    "Authorization": f"Bearer {GITHUB_TOKEN}",
    "Accept": "application/vnd.github+json",
}

user_resp = requests.get("https://api.github.com/user", headers=gh_headers)
user_resp.raise_for_status()
gh_user = user_resp.json()

repos_resp = requests.get("https://api.github.com/user/repos", headers=gh_headers)
repos_resp.raise_for_status()
repos = repos_resp.json()

print(f"GitHub user: {gh_user['login']}")
print(f"Total repos: {len(repos)}")

# Collect last commit message for first few repos
for repo in repos[:3]:
    commits_url = f"https://api.github.com/repos/{gh_user['login']}/{repo['name']}/commits"
    commits_resp = requests.get(commits_url, headers=gh_headers)
    commits_resp.raise_for_status()
    commits = commits_resp.json()
    last_msg = commits[0]["commit"]["message"] if commits else "No commits"
    print(f"- {repo['name']}: {last_msg}")

# ---------- Other website ----------
session = requests.Session()

login_url = "https://example.com/login"
protected_url = "https://example.com/api/profile"

login_payload = {
    "username": os.environ.get("APP_USER"),
    "password": os.environ.get("APP_PASS"),
}

login_resp = session.post(login_url, data=login_payload)
login_resp.raise_for_status()

profile_resp = session.get(protected_url)
profile_resp.raise_for_status()
profile = profile_resp.json()

print("\nExternal app profile:")
print(profile)
```
